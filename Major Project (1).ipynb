```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(df)
pred=kmeans.predict(df)
```

```python
df['cluster'] = pred
```

```python
def seg(str_x, str_y, clusters):
  x = []
  y = []
  for i in range(clusters):
    x.append(df[str_x][df['cluster']==i])
    y.append(df[str_y][df['cluster']==i])
  return x,y
def plot_clusters(str_x, str_y, clusters):
  plt.figure(figsize = (5,5), dpi = 120)
  x,y = seg(str_x, str_y, clusters)
  for i in range(clusters):
    plt.scatter(x[i], y[i], label = 'cluster{}'.format(i))
  plt.xlabel(str_x)
  plt.ylabel(str_y)
  plt.title(str(str_x+" Vs "+str_y))
  plt.legend()
```

```python
plot_clusters('total_price', 'Quantity',3)
```
